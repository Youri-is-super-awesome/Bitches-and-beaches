```python

def preprosses(df, target, train_size = 0.7):
    
    """convert the pandas dataframes to numpy ndarrays"""
    
    X_np = df.to_numpy()
    y_np = target.to_numpy()
    
    #split the data into 70% training and 30% testing
    X_train, X_test, y_train, y_test = train_test_split(X_np, y_np, train_size=train_size, random_state=1265599650)
    
    return X_train, X_test, y_train, y_test

    

    
    
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error



def calc_mae_rmse(X_train, X_test, y_train, y_test, linear_model):
    
    
    linear_model.fit(X_train, y_train)
    
    predict = linear_model.predict(X_test)

    linear_test_mae = mean_absolute_error(y_test, predict)
    linear_test_rmse = np.sqrt(mean_squared_error(y_test, predict))
    
    print(f'The test MAE for our multivariate linear model {linear_test_mae}')
    print(f'The test RMSE for our multivariate linear model {linear_test_rmse}')

    #print(f"The theta values for our multivariate linear model are:\n{linear_model.intercept_} and {linear_model.coef_}.")





#linear_model2.fit(X_train, y_train)

# predict = linear_model2.predict(X_test)

# linear_test_mae = mean_absolute_error(y_test, predict)
# linear_test_rmse = np.sqrt(mean_squared_error(y_test, predict))

# print(f'The test MAE for our linear model {linear_test_mae}')
# print(f'The test RMSE for our linear model {linear_test_rmse}')

# print(f"The theta values for the sklearn linear model are:\n{linear_model2.intercept_} and {linear_model2.coef_}.")
```

#### Polynomial features 

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# X = X_train
# X_ = X_train_
# vector = y_train
# predict = X_test
# predict_ = X_test_




def add_poly(X_train, X_test, y_train, y_test, poly_model, deg = 2):
    
    # change feature matrix into polynomial feature matrix
    poly = PolynomialFeatures(degree = deg)
    X_train_ = poly.fit_transform(X_train)
    X_test_ = poly.fit_transform(X_test)
    


    poly_model.fit(X_train_, y_train)

    predictions = poly_model.predict(X_test_)

    poly_test_mae = mean_absolute_error(y_test, predictions)
    poly_test_rmse = np.sqrt(mean_squared_error(y_test, predictions))

    print(f'The test MAE for our polynomial linear model {poly_test_mae}')
    print(f'The test RMSE for our polynomial linear model {poly_test_rmse}')

    #print(f'The theta values for our polynomial linear model are:\n{poly_model.coef_} and {poly_model.intercept_}')



# change feature matrix into polynomial feature matrix
# poly = PolynomialFeatures(degree=2)
# X_train_ = poly.fit_transform(X_train)
# X_test_ = poly.fit_transform(X_test)

# # implement model
# poly_model = LinearRegression()

# poly_model.fit(X_train_, y_train)

# predictions = poly_model.predict(X_test_)

# poly_test_mae = mean_absolute_error(y_test, predictions)
# poly_test_rmse = np.sqrt(mean_squared_error(y_test, predictions))

# print(f'The test MAE for our linear model {poly_test_mae}')
# print(f'The test RMSE for our linear model {poly_test_rmse}')

# print(f'The theta values for the sklearn linear model are:\n{poly_model.coef_} and {poly_model.intercept_}')
```

#### Only interaction features 

```python
# change feature matrix into polynomial feature matrix



def poly_interact(X_train, X_test, y_train, y_test, poly_model_interact):

    poly_interact = PolynomialFeatures(interaction_only=True)
    X_train_interact = poly_interact.fit_transform(X_train)
    X_test_interact = poly_interact.fit_transform(X_test)
    
    poly_model_interact.fit(X_train_interact, y_train)

    predictions = poly_model_interact.predict(X_test_interact)

    poly_test_mae = mean_absolute_error(y_test, predictions)
    poly_test_rmse = np.sqrt(mean_squared_error(y_test, predictions))

    print(f'The test MAE for our interact linear model {poly_test_mae}')
    print(f'The test RMSE for our interact linear model {poly_test_rmse}')

    #print(f'The theta values for the interact linear model are:\n{poly_model_interact.coef_} and {poly_model_interact.intercept_}')


    
```
